```python
import pandas as pd
import re
```

```python
def text_cleaning(x):
    
    return x.replace(' ','').replace('(','').replace(')','').replace('[0-9]+','')
```

```python
movie_meta_df = pd.read_excel('../../영화정보 리스트_2018-05-24.xls')
movie_meta_df.info()
```

```python
train_df = pd.read_csv('../../TBC_MOVIES_TITLE.csv',encoding='euc-kr')
train_df.info()
```

```python
train_df['TITLE'] = train_df['TITLE'].apply(text_cleaning)
movie_meta_df['영화명'] = movie_meta_df['영화명'].apply(text_cleaning)
movie_meta_df = movie_meta_df.sort_values('제작연도', ascending=False).drop_duplicates('영화명').sort_index()
```

```python
result = train_df.merge(movie_meta_df, left_on='TITLE', right_on='영화명', how='left')
result.info()
```
